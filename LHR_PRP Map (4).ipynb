```python
import folium

hackensack_coords = [40.738893, -74.104844]


m = folium.Map(location=hackensack_coords, zoom_start=13, tiles=None)

folium.TileLayer(
    tiles='https://server.arcgisonline.com/ArcGIS/rest/services/World_Imagery/MapServer/tile/{z}/{y}/{x}',
    attr='ESRI World Imagery',
    name='Satellite',
    overlay=False,
    control=True
).add_to(m)
```

```python
locations1 = [
    {"name": "Western Electric Site (Nokia)", "lat": 40.73149253278342, "lon": -74.11383937631429},
    {"name": "Standard Chlorine Site (Beazer,OxyChem, Cooper, Apogent)", "lat": 40.75041765205132, "lon": -74.09951944232327},
    {"name": "Federal Shipbuilding Site (US Steel, RTC Properties)", "lat": 40.724401, "lon": -74.106595},
    {"name": "Meadows Rail Yard Facility (Conrail)", "lat": 40.74429235539165, "lon": -74.09892594747733},
    {"name": "United Cork/BASF Site (BASF)", "lat": 40.72030593141203, "lon": -74.11614328795092}]

radius=10
for loc in locations1:
    folium.CircleMarker(
        location=[loc["lat"], loc["lon"]],
        radius=radius,
        color="black",
        weight=1,
        fill_opacity=0.6,
        opacity=1,
        fill_color="orange",
        fill=False,
        popup="{} pixels".format(radius),
        tooltip=loc["name"]
    ).add_to(m)

locations2 = [
    {"name": "PJP Landfill", "lat": 40.735198082066, "lon": -74.08127133027752},
    {"name": "Nicholas Galvanizing Site", "lat": 40.74331668648509, "lon": -74.07381394376795},
    {"name": "Kuehne Chemical Site", "lat": 40.734152336390444, "lon": -74.1005129186414},
    {"name": "PMC Global Site", "lat": 40.76413159448509, "lon": -74.13998306211818},
    {"name": "Trumbull Asphalt Site", "lat": 40.74438574017153, "lon": -74.0871082898045}]

for loc in locations2:
    folium.CircleMarker(
        location=[loc["lat"], loc["lon"]],
        radius=radius,
        color="black",
        weight=1,
        fill_opacity=0.6,
        opacity=1,
        fill_color="yellow",
        fill=False,
        popup="{} pixels".format(radius),
        tooltip=loc["name"]
    ).add_to(m)

m
```
